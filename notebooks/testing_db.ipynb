```python
import sqlite3
from pathlib import Path

db = Path("/Users/lau/workspace/personal/ocean_ai/dory-bot/data/dory.db")
connector = sqlite3.connect(db)
cursor    = connector.cursor()

all_table_names_query = "SELECT name FROM sqlite_master WHERE type='table'"

cursor.execute(all_table_names_query)

output = cursor.fetchall()

table_names = []

for table in output:
    table_names.append(table[0])
```

```python
# Inspecting each table

for table_name in table_names:
    table_inspection_query = f"SELECT * FROM {table_name}"
    result = cursor.execute(table_inspection_query)
    print(cursor.fetchall())
    # print(table_inspection_query)
```
