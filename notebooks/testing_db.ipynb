```python
import sqlite3
from pathlib import Path

path = Path("../data") / "kato.db"
```

The `connect()` method of sqlite3 allows you to interface with your database through python

```python
connector = sqlite3.connect(path)
```

Once you've connected, you navigate using the `cursor()` method

```python
cursor = connector.cursor()
```

SQL = programming in plain English

Keywords:
- SELECT: is like a filter, this is where you tell the program, which columns to retrieve from your data
- FROM: is the filter for the table where your data lives
- WHERE: is a condition that you can apply. Like boolean logic e.g. WHERE sales_value > 10
- JOIN:
- CREATE:
- ...
